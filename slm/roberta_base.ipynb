```python
import torch
def predict(sentence):
  token_ids = tokenizer.convert_tokens_to_ids(tokenizer.tokenize(sentence))
  input_tensor = torch.tensor([token_ids])
  outputs = model(input_tensor)
  result = torch.argmax(outputs[0]).item()

  return 'positive' if result == 1 else 'negative'

from adapters import AutoAdapterModel
from transformers import AutoTokenizer
model = AutoAdapterModel.from_pretrained(f"D:/ds/work/workspace/git/my_model_repo/models/roberta-base")
tokenizer = AutoTokenizer.from_pretrained(f"D:/ds/work/workspace/git/my_model_repo/models/roberta-base")
adapter_name = model.load_adapter("AdapterHub/roberta-base-pf-sst2")
model.set_active_adapters(adapter_name)

print(predict("we will win this game"))
```
